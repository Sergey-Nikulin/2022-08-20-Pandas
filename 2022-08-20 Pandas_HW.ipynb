#### Задание 1
Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
films = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
# films.head()
# ratings.head()
joined = ratings.merge(films, on='movieId', how='left')
joined.head()
# joined.shape
```

```python
filtr_films = joined[ (joined.rating == int(5)) ]

filtr_films.title.value_counts().head(1)
```

#### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
# data.head()
# data.info()
```

```python
filtered_countries = data[ ((data['country'] == 'Latvia') | (data['country'] == 'Estonia') | (data['country'] == 'Lithuania'))\
& ((data['category'] == int(4)) | (data['category'] == int(12)) | (data['category'] == int(21)))\
& ((data['year'] >= int(2005)) & (data['year'] <= int(2010)))\
& ((data['quantity'] > int(0))) ]

filtered_countries.sort_values(by=['country', 'year', 'quantity'], ascending=[True, True, True]).head()
```

```python
# filtered_countries.info()
```

```python
# суммарное потребление стран Прибалтики:
sum_potr = filtered_countries.quantity.sum()
sum_potr
```
